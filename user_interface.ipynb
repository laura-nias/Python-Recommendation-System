# User Interface

Runs the code for the interface. It provides the user with options and the user can then enter the relevant number or string. 

```python
import similarity_module as sim
import load_dataset_module as load

def main():
    
    def dict_choice():
        print("\nOPTIONS\n")
        print("1. Compare Artist Features")
        print("2. Compare Music Features")
        print("3. Quit Application")

        #Lets user input a choice for which dictionary to use or to quit the application
        try:
            #Strips input of leading and trailing whitespace
            choice = input("Choose an option (Enter corresponding number): ").strip()
            #Turns string into int
            choice = int(choice)
            
            if(choice == 1):
                dict_1 = load.artist()
                enter_id(dict_1)
            elif(choice == 2):
                dict_1 = load.music()
                enter_id(dict_1)
            elif(choice == 3):
                return
            elif(choice > 3): 
                #Triggers an error if the value entered is not in range
                print("ERROR: Value is not in range")
                dict_choice()

        #Triggers an error if the value entered is not a number
        except ValueError:
            print("ERROR: Input was not an integer")
            dict_choice()
        #Handles any other error
        except:
            print("ERROR: Something went wrong")
            dict_choice()
    
    def enter_id(dict_1):
        try:
             #Strips input of leading and trailing whitespace
            id_1 = input("\nEnter the first ID you wish to compare: ").strip()
            id_2 = input("Enter the second ID you wish to compare: ").strip()
            
            #If IDs are not the same...
            if(id_1 != id_2):
                 #If IDs are valid...
                if(id_1 in dict_1 and id_2 in dict_1):
                    metric_choice(dict_1, id_1, id_2)
                elif(id_1 not in dict_1):
                    #Triggers an error if the ID entered is not valid
                    print("ERROR: First ID is not valid")
                    if(id_2 not in dict_1):
                        print("ERROR: Second ID is not valid")
                    enter_id(dict_1)
                elif(id_2 not in dict_1):
                    #Triggers an error if both of the IDs entered are not valid
                    print("ERROR: Second ID is not valid")
                    enter_id(dict_1)
            else:
                #Triggers an error if both of the IDs entered are the same
                print("Those two IDs are the same, please enter different ones")
                enter_id(dict_1)
                    
        except:
            #Handles any other error
            print("ERROR: Something went wrong")
            enter_id(dict_1)
    
    def metric_choice(dict_1, id_1, id_2):
        print("\nMETRICS\n")
        print("1. Euclidean")
        print("2. Cosine")
        print("3. Pearson")
        print("4. Jaccard")
        print("5. Manhattan")
        print("6. Quit Application")

        #Lets user input a choice for which metric to use or to quit the application
        try:
            #Strips input of leading and trailing whitespace
            metric = input("Which metric would you like to use? (Enter corresponding number): ").strip()
            #Turns string into int
            metric = int(metric)
            
            if(metric == 6):
                return
            elif(metric == 1):
                print("Euclidean Similarity: ", sim.euclidean_similarity(dict_1, id_1, id_2))
            elif(metric == 2):
                print("Cosine Similarity: ", sim.cosine_similarity(dict_1, id_1, id_2))
            elif(metric == 3):
                print("Pearson Similarity: ", sim.pearson_similarity(dict_1, id_1, id_2))
            elif(metric == 4):
                print("Jaccard Similarity: ", sim.jaccard_similarity(dict_1, id_1, id_2))
            elif(metric == 5):
                print("Manhattan Similarity: ", sim.manhattan_similarity(dict_1, id_1, id_2))
            else: 
                #Triggers an error if the value entered is not in range
                print("ERROR: Value is not in range")
                metric_choice(dict_1, id_1, id_2)

        #Triggers an error if the value entered is not a number
        except ValueError:
            print("ERROR: Input was not an integer")
            metric_choice(dict_1, id_1, id_2)
        #Handles any other error
        except:
            print("Something went wrong")
            metric_choice(dict_1, id_1, id_2)
            
    dict_choice()
    
```

```python
#Calls the main function
main()
```
